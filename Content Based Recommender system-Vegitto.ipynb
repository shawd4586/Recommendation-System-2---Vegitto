```python
for k in range(len(d)):
    for i in range(len(d[k])-1,-1,-1):
        for j in ingredients.keys():
            for x in ingredients[j]:
                if x in d[k][i]:
                    d[k][i]=j
        if d[k][i] not in ingredients.keys():
            del d[k][i]
```

```python
for i in range(d.shape[0]):
    data['مواد اولیه'][i]=d[i]
```

```python
icm=data.copy()
for index, row in data.iterrows():
    for ingred in row['مواد اولیه']:
        icm.at[index, ingred] = 1
icm=icm.fillna(0)
icm.drop(['نام غذا','مواد اولیه'],axis=1,inplace=True)
icm.rename(columns={'کالری برای یک نفر':'calorie',
                   'سریع':'time',
                   'وعده غذایی':'meal',
                   'نوع رژیم گیاهخواری':'type',
                   'کشور':'region'},inplace=True)
```

```python
userfeature=icm[icm['ID'].isin(random_user['id'].tolist())]
userfeature2=icm[icm['ID'].isin(random_user2['id'].tolist())]
userfeature3=icm[icm['ID'].isin(random_user3['id'].tolist())]
userfeature=userfeature.reset_index(drop=True).drop('ID',1)
userfeature2=userfeature2.reset_index(drop=True).drop('ID',1)
userfeature3=userfeature3.reset_index(drop=True).drop('ID',1)
```

```python
userfeature2
```

```python
userProfile=userfeature.transpose().dot(random_user['point'])
userProfile2=userfeature2.transpose().dot(random_user2['point'])
userProfile3=userfeature3.transpose().dot(random_user3['point'])
```

```python
userProfile
```

```python
recommendation=(((icm.drop(icm[icm['ID'].isin(random_user['id'].tolist())].index).set_index('ID')*userProfile).sum(1)/(userProfile.sum())).sort_values(ascending=False)).reset_index(drop=False)
recommendation2=(((icm.drop(icm[icm['ID'].isin(random_user2['id'].tolist())].index).set_index('ID')*userProfile2).sum(1)/(userProfile2.sum())).sort_values(ascending=False)).reset_index(drop=False)
recommendation3=(((icm.drop(icm[icm['ID'].isin(random_user3['id'].tolist())].index).set_index('ID')*userProfile3).sum(1)/(userProfile3.sum())).sort_values(ascending=False)).reset_index(drop=False)
```
